# Matplot

```python
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tn = sns.load_dataset("titanic")
```

```python
tn
```

### Scatterplot using plt 

```python
plt.scatter(tn["age"], tn["fare"], color = "red", alpha = 0.5)
plt.xlabel("age")
plt.ylabel("fare")
plt.title("Representing relation between age and fare")
```

### Lineplot using plt 

```python
plt.plot(tn["age"], tn["fare"])
plt.xlabel("age")
plt.ylabel("fare")
plt.title("Representing relation between age and fare")
```

### Barplot using plt

```python
plt.bar(tn["age"], tn["fare"])
plt.xlabel("age")
plt.ylabel("fare")
plt.title("fare over age")
```

```python
plt.hist(tn["age"], bins=20, color="blue", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution of Passengers")
```

```python
plt.hist(tn["fare"].dropna(), bins=20, color="green", edgecolor="black")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.title("Fare Distribution of Passengers")
```
