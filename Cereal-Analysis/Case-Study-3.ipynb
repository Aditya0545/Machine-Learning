### 3. Extract the rating as your target variable ‘y’ and all numerical parameters as your predictors ‘x’. Separate 25% of your data as test set.

```python
X = df.drop(["rating", "name", "mfr", "type", "manufacturerNames"], axis=1)
Y = df["rating"]
```

```python
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=85)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.figure(figsize=(12,8))
colors = np.where(y_pred>y_test,'red','blue')
plt.scatter(y_test, y_pred, c = colors)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual VS Predicted")
```
